```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
x=[0,1,2,3,4]
y = [0,2,4,6,8]
#plt.plot(x,y,label='y=mx+c',color='red',linewidth='3',marker='x',markersize='10',markeredgecolor='blue',linestyle='--')
#resize your graph
plt.figure(figsize=(5,3),dpi=100)
#shorthand
# fmt = [color][marker][line]
#line 1
plt.plot(x,y,'bX--',label='2x',linewidth='3',markersize='10',markeredgecolor='blue')

#line 2
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:],'r--')

plt.title('Graph Title',fontdict={'fontname' : 'Comic Sans MS','fontsize':20})
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks([1,2,3,4])
#plt.yticks([2,4,6,8,20])
plt.legend()
plt.savefig('mygraph.png',dpi=300)
plt.show()
```

# Barchart

```python
labels = ['A','B','c']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['/','o','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')
plt.figure(figsize=(6,4))
plt.show()
```
